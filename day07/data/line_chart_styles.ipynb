## Exercise 01 : Line chart with styles

#### Import and connection

```python
import pandas as pd
import sqlite3

try:
  connect = sqlite3.connect('../data/checking-logs.sqlite.sqlite')
  print('OK. Connected to sqlite')

except sqlite3.Error as error:
  print('Error with connection:', error)
```

#### Get only the users and not the admins (group by date) from the pageviews table

```python
query = '''
SELECT uid, datetime
FROM pageviews
WHERE uid LIKE 'user_%'
'''
views = pd.io.sql.read_sql(query, connect, parse_dates=['datetime'])
views = views.groupby(by=views['datetime'].dt.date).count()['uid'].reset_index()
views.columns = ['date', 'views']
views = views.set_index('date')
views
```

#### Get only the users and not the admins (group by date) from the checker table

```python
query = '''
SELECT uid, timestamp
FROM checker
WHERE uid LIKE 'user_%'
'''
commits = pd.io.sql.read_sql(query, connect, parse_dates=['timestamp'])
commits = commits.groupby(by=commits['timestamp'].dt.date).count()['uid'].reset_index()
commits.columns = ['date', 'commits']
commits = commits.set_index('date')
commits
```

#### Analyze only the dates when there were both views and checker commits

```python
views_commits = pd.concat([views, commits], axis=1)
views_commits = views_commits.dropna()
views_commits
```

```python
views_commits.info()
```

####  Create a graph with Pandas method .plot()

```python
ax = views_commits.plot(fontsize=8,
                        figsize=(15,8),
                        title='Views and commits per day',
                        color=['red', 'c'], style=['-.', '--'], rot=90)
ax.grid(axis='x')
ax.get_lines()[0].set_marker('o')
```

#### How many times was the number of views larger than 150?

```python
views_commits[views_commits['views'] > 150]['views'].count()
```

### The answer is 2
