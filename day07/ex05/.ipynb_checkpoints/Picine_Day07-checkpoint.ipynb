```python
import pandas as pd
import sqlite3
```

```python
from google.colab import drive  # если вы выполняете код из среды Google Colab, нужно подключить свой гугл-диск,
drive.mount('/content/drive')
```

```python
connect = sqlite3.connect('/content/drive/MyDrive/Pyton/checking-logs.sqlite.sqlite copy')
```

Подготовим таблицу просмотров страниц, выбирая только пользователей, а не администраторов и группируя по дням

```python
query = '''
SELECT uid, datetime
FROM pageviews
WHERE uid LIKE 'user_%'
'''
views = pd.io.sql.read_sql(query, connect, parse_dates=['datetime'])
views = views.groupby(by=views['datetime'].dt.date).count()['uid'].reset_index()
views.columns = ['date', 'views']
views = views.set_index('date')
views
```

Подготовим таблицу коммитов страниц, выбирая только пользователей, а не администраторов и группируя по дням

```python
query = '''
SELECT uid, timestamp
FROM checker
WHERE uid LIKE 'user_%'
'''
commits = pd.io.sql.read_sql(query, connect, parse_dates=['timestamp'])
commits = commits.groupby(by=commits['timestamp'].dt.date).count()['uid'].reset_index()
commits.columns = ['date', 'commits']
commits = commits.set_index('date')
commits
```

Объединим полученные таблицы в 

Так как нужны только даты, когда были как просмотры, так и коммиты, удалим строки с пустыми значениями (dropna)

```python
views_commits = pd.concat([views, commits], axis=1)
views_commits = views_commits.dropna()
views_commits
```

Построим график с заданными характеристиками

https://pyprog.pro/mpl/mpl_main_components.html

```python
ax = views_commits.plot(fontsize=8,
                        figsize=(15,8),
                        title='Views and commits per day',
                        color=['red', 'c'], style=['-.', '--'], rot=90)
ax.grid(axis='x') # сетка вертикальными линиями
ax.get_lines()[0].set_marker('o') # точки на красном графике
```

Посмотрим, сколько раз количество просмотров превышало 150

```python
views_commits[views_commits['views'] > 150]['views'].count()
```

How many times was the number of views larger than 150?

The answer is 2.

```python
connect.close()
```

Подготовим таблицу. Выберем из datetime отдельно дату и время

```python
query = '''
SELECT uid, timestamp
FROM checker
WHERE uid LIKE 'user_%'
'''
commits = pd.io.sql.read_sql(query, connect, parse_dates=['timestamp'])
commits['date'] = commits.timestamp.dt.date
commits['hour'] = commits.timestamp.dt.hour
commits
```

Добавим колонку с названием времени суток

```python
commits['daytime'] = pd.cut(commits.hour,
                            bins=[0, 4, 10, 17, 24],
                            labels=['nigth', 'morning',
                                    'afternoon', 'evening'],
                            right=False,
                            include_lowest=True)
commits
```

```python
commits_group = commits.groupby('daytime').date.value_counts().unstack(0)
commits_group
```

```python
ax = commits_group.plot(kind='bar',
                   stacked=True,
                   figsize=(15, 8),
                   title='Commits per day',
                   fontsize=8,
                   rot=90,
                   color=['#00008b', '#fffb00', '#ff5500', '#680080'])
ax.set_facecolor('#c2fff4') # цвет фона
ax.grid('on', which='major', axis='y') # горизонтальная сетка
```

Посмотрим суммарное количество коммитов в определенное время суток

```python
commits_group.sum()
```

When do our users usually commit the labs: in the night, morning, afternoon,

The answer is evening and afternoon.

```python
import pandas as pd
import sqlite3
```

```python
from google.colab import drive  # если вы выполняете код из среды Google Colab, нужно подключить свой гугл-диск,
drive.mount('/content/drive')
```

```python
connect = sqlite3.connect('/content/drive/MyDrive/Pyton/checking-logs.sqlite.sqlite copy')
```

Подготовим таблицу. Выберем из datetime отдельно день недели и время коммита, определим, является ли день выходным и уберем лишнее

```python
query = '''
SELECT uid, timestamp
FROM checker
WHERE uid LIKE 'user_%'
'''
commits = pd.io.sql.read_sql(query, connect, parse_dates=['timestamp'])
commits['weekday'] = commits['timestamp'].dt.dayofweek
commits['day'] = commits.timestamp.dt.day
commits['hour'] = commits.timestamp.dt.hour
commits['type'] = commits['weekday'].apply(
    lambda x: 'working_day' if x < 5 else 'weekend')
commits = commits.drop(columns = ['weekday'])
commits
```

Для каждого часа рассчитаем среднее количество коммитов в рабочие дни и в
выходные дни (если за час коммитов не было, не используем его для вычисления среднего значения) 

```python
commits_groupped = commits.groupby(['type', 'day', 'hour']).count()
commits_groupped = commits_groupped.groupby(
    ['type', 'hour']).mean()['uid'].reset_index()
commits_groupped
```

Создадим сводную таблицу (pd.pivot)

https://dfedorov.spb.ru/pandas/Сводная%20таблица%20в%20pandas.html

```python
pivot_table = pd.pivot(commits_groupped,
                       values='uid',
                       index='hour',
                       columns='type')
pivot_table
```

Построим график

https://pyprog.pro/mpl/mpl_main_components.html

```python
ax = pivot_table.plot(kind='bar',
                      figsize=(15, 8),
                      title='Commits per hour',
                      rot=0,
                      fontsize=8,
                      color=['green', 'blue'],
                      subplots=True) # False поместило бы гистограммы рядом (в один график)
```

Посмотрим, в какое время наибольшее количество коммитов в рабочие дни и выходные

```python
pivot_table[pivot_table['working_day'] == pivot_table['working_day'].max()]
```

```python
pivot_table[pivot_table['weekend'] == pivot_table['weekend'].max()]
```

Вывод: динамика в рабочие и выходные дни отличается. В рабочие дни пик коммитов смещен на вечер, а в выходные на утро.

```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
```

```python
from google.colab import drive  # если вы выполняете код из среды Google Colab, нужно подключить свой гугл-диск,
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Pyton/ab-test.csv')
df.head()
```

Построим график с указанными требованиями

boxplot https://pythobyte.com/matplotlib-boxplot-3bc87379/

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 8)) # 1 кол-во частей по вертикали, 2 по горизонтали
df[df['time'] == 'after'].boxplot(column='diff', # значения из колонки 'diff'
                                  by='group', # сгруппированы по значению 'group'
                                  ax=ax[0], # расположен в 0 секторе
                                  color=dict(boxes='#009e10', whiskers='red',
                                             medians='red', caps='red'), # цвета всех линий
                                  boxprops=dict(linewidth=3), # изменение толщины линии коробки
                                  medianprops=dict(linewidth=2)) # изменение толщины линии медианы
df[df['time'] == 'before'].boxplot(column='diff',
                                   by='group',
                                   ax=ax[1],
                                   color=dict(boxes='green', whiskers='red',
                                              medians='red', caps='red'),
                                   boxprops=dict(linewidth=3),
                                   medianprops=dict(linewidth=2))
ax[0].set_title('after')
ax[1].set_title('before')
fig.suptitle(
    'Lag before labs deadlines. After and before the Newsfeed', fontsize=15)
plt.show()
```

Интерквартильный размах (IQR) — число, которое показывает разброс средней половины (т.е. средние 50%) набора данных и помогает определить выбросы

https://pythobyte.com/detection-removal-outliers-in-python-adaa857e/

```python
iqr = df[(df['time'] == 'before') & (df['group'] == 'control')].describe().loc['75%'].values[0] - \
    df[(df['time'] == 'before') & (df['group'] == 'control')
       ].describe().loc['25%'].values[0]
round(iqr / 10) * 10
```

```python
iqr = df[(df['time'] == 'after') & (df['group'] == 'control')].describe().loc['75%'].values[0] - \
    df[(df['time'] == 'after') & (df['group'] == 'control')
       ].describe().loc['25%'].values[0]
round(iqr / 10) * 10
```

Вывод. IQR контрольной группы до появления ленты новостей примерно 150, а после появления ленты новостей примерно 70, что говорит о существенном уменьшении разброса значений

```python
import pandas as pd
import sqlite3
from pandas.plotting import scatter_matrix
```

```python
from google.colab import drive  # если вы выполняете код из среды Google Colab, нужно подключить свой гугл-диск,
drive.mount('/content/drive')
```

```python
connect = sqlite3.connect('/content/drive/MyDrive/Pyton/checking-logs.sqlite.sqlite copy')
```

Посмотрим, какие таблицы есть в базе test и control это таблицы, которые мы сами создали (выборки)

```python
a = connect.cursor().execute('SELECT name from sqlite_master where type= "table"')
print(a.fetchall())
```

Создадим таблицу с количеством коммитов

```python
query = '''
SELECT uid, COUNT(*) AS num_commits
FROM checker
WHERE uid LIKE 'user_%'
AND labname <> 'project1'
GROUP BY uid
'''
commits = pd.io.sql.read_sql(query, connect)
commits
```

Создадим таблицу с количеством просмотров

```python
query = '''
SELECT uid, COUNT(*) AS num_views
FROM pageviews
WHERE uid LIKE 'user_%'
GROUP BY uid
'''
num_views = pd.io.sql.read_sql(query, connect)
num_views
```

Создадим таблицу с дельтой: разница между первым коммитом и дедлайном лаборатории (подробно, как мы это делали в Day06 ех03).  (Не принимать во внимание project1 для расчета средней разницы и количества коммитов)

```python
query = '''
SELECT uid,
       CAST((JulianDay(test.first_commit_ts) -
            JulianDay(DATETIME(deadlines.deadlines, 'unixepoch'))
                  ) * 24 AS Integer
           ) AS delta
FROM test
LEFT JOIN deadlines ON test.labname=deadlines.labs
WHERE labname <> 'project1'
'''
delta = pd.io.sql.read_sql(query, connect)
delta
```

Найдем среднюю дельту по каждому юзеру

```python
delta = delta.groupby(['uid']).mean()
```

Объединим все это в один датафрейм, в котором каждый пользователь тестовой группы имеет среднюю разницу между первым коммитом и дедлайном лаборатории, количество просмотров страниц и количество коммитов (не принимать во внимание project1 для расчета средней разницы и количества коммитов)

```python
pivot_table = num_views.merge(commits, on=['uid'])
pivot_table = delta.merge(pivot_table, on=['uid'])
pivot_table = pivot_table.drop(columns = ['uid'])
pivot_table
```

По диагоналям отображена плотность ядра (KDE). KDE - это непараметрический способ оценки функции плотности вероятности любой переменной.

```python
ax = scatter_matrix(pivot_table,
                    figsize=(15, 8),
                    diagonal='kde',
                    density_kwds=dict(color='#059ABF', linewidth=3),
                    s=200,
                    color='red')
```

```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

Создадим таблицу с юзерами и датой-временем коммита

```python
query = '''
SELECT uid, timestamp
FROM checker
WHERE uid LIKE 'user_%'
'''
commits = pd.io.sql.read_sql(query, connect, parse_dates=['timestamp'])
```

Переделаем таблицу на таблицу с днями недели и временем

```python
days = {0: 'Mon', 1: 'Tue', 2: 'Wed', 3: 'Thu', 4: 'Fri', 5: 'Sat', 6: 'Sun'}

commits['weekday'] = commits['timestamp'].apply(lambda x: days[x.dayofweek])
commits['hour'] = commits['timestamp'].dt.hour
commits = commits.drop('timestamp', axis=1)
commits
```

Создадим таблицу, в которой разложим коммиты юзеров по дням недели

Отсортируем имена юзеров по уменьшению общего количества коммитов, чтобы использовать их для названий колонок

```python
columns = commits.groupby('uid').count().sort_values('hour', ascending=False).index
columns
```

```python
df_1 = commits.groupby('uid').weekday.value_counts().unstack(0) # unstack переворачивает таблицу
df_1 = df_1.reindex(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']) # отсортируем дни по порядку
df_1 = df_1.reindex(columns=columns)# отсортируем юзеров по уменьшению коммитов согласно нашего подготовленного списка
df_1 = df_1.fillna(0) # пустые значения заполним нулем
df_1
```

Создадим таблицу, в которой разложим коммиты юзеров по часам

Для индекса будем использовать числа от 0 до 23 (правая граница stop=24 не входит в диапазон) 

```python
index = pd.RangeIndex(stop=24)
```

```python
df_2 = commits.groupby('uid').hour.value_counts().unstack(0)
df_2 = df_2.reindex(index=index, fill_value=0)
df_2 = df_2.reindex(columns=columns)
df_2 = df_2.rename_axis('hour')
df_2 = df_2.fillna(0)
df_2
```

Построим первый график

```python
fig, ax = plt.subplots(figsize=(15, 8))
im = ax.imshow(df_1, cmap='RdBu') # imshow расположение координат
divider = make_axes_locatable(ax) # колонка градиента для закраски
cax = divider.append_axes('right', size='5%', pad=0.05) # ее расположение, иразмер и градация
plt.colorbar(im, cax=cax) # эту же цветовую градацию используем для графика
ax.set_xticks(range(df_1.shape[1])) # определяем, какие значения на каких осях
ax.set_yticks(range(df_1.shape[0]))
ax.set_xticklabels(df_1.columns) # определяем, как подписаны оси
ax.set_yticklabels(df_1.index)
plt.setp(ax.get_xticklabels(), rotation=90) # поворачиваем подписи на оси х на 90 градусов
plt.show()
```

Построим второй график

```python
fig, ax = plt.subplots(figsize=(15, 8))
im = ax.imshow(df_2, cmap='jet') # imshow расположение координат
divider = make_axes_locatable(ax) # колонка градиента для закраски
cax = divider.append_axes('right', size='5%', pad=0.05) # ее расположение, иразмер и градация
plt.colorbar(im, cax=cax) # эту же цветовую градацию используем для графика
ax.set_xticks(range(df_2.shape[1])) # определяем, какие значения на каких осях
ax.set_yticks(range(df_2.shape[0]))
ax.set_xticklabels(df_2.columns) # определяем, как подписаны оси
ax.set_yticklabels(df_2.index)
plt.setp(ax.get_xticklabels(), rotation=90) # поворачиваем подписи на оси х на 90 градусов
plt.show()
```

У какого пользователя больше всего коммитов во вторник?

```python
import sqlite3
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials
```

```python
connect = sqlite3.connect('/content/drive/MyDrive/Pyton/checking-logs.sqlite.sqlite copy')
```

```python
conn = sqlite3.connect('data/checking-logs.sqlite')
```

```python
commits = pd.io.sql.read_sql(
    """
    select uid, numTrials, timestamp as date
    from checker
    where uid like 'user_%' and status = 'ready' and labname = 'project1'
    """,
    connect, parse_dates=['date']
    )
commits['date'] = commits['date'].dt.floor('d')
```

```python
commits.head()
```

```python
connect.close()
```

```python
df = commits.groupby(['date', 'uid']).max()['numTrials'].unstack()
df.index = np.arange(len(df.index))
df.iloc[0] = df.iloc[0].fillna(0)
df = df.fillna(method='pad', axis=0)
df = df.transpose()
df
```

```python
initial_data = [go.Scatter(x=np.array([]), y=np.array([]),
                           mode='lines+markers', name=name)
                for name in df.index]
```

```python
frames = [go.Frame(data=[go.Scatter(
    x=np.arange(i + 1), y=np.array(df.iloc[j, :i + 1]),
    mode='lines+markers', name=df.index[j]
    ) for j in range(len(df.index))]) for i in range(len(df.columns) + 1)]
```

```python
fig = go.Figure(
    data=initial_data,
    layout={
        'width': 1200,
        'height': 600,
        'title': 'Dynamic of commits per user in project1',
        'xaxis': {'range': (0, len(df.columns) + 1), 'dtick': 2},
        'yaxis': {'range': (0, round(max(df.max()) + 2, -1) + 1)},
        'updatemenus': [{
            'type': 'buttons',
            'buttons': [{'method': 'animate', 'label': 'play', 'args': [None]}]
            }]}, frames=frames)
fig.show()
```
