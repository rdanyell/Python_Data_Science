### Import and connection

```python
import pandas as pd
import sqlite3

try:
  connect = sqlite3.connect('../data/checking-logs.sqlite')
  print('OK. Connected to sqlite')

except sqlite3.Error as error:
  print('Error with connection:', error)
```

#### Get 'date' and 'hour' from the checker table for only the users and not the admins

```python
query = '''
SELECT uid, timestamp
FROM checker
WHERE uid LIKE 'user_%'
'''
commits = pd.io.sql.read_sql(query, connect, parse_dates=['timestamp'])
commits['date'] = commits.timestamp.dt.date
commits['hour'] = commits.timestamp.dt.hour
commits
```

#### Add 'daytime' column to the dataframe

```python
commits['daytime'] = pd.cut(commits.hour,
                            bins=[0, 4, 10, 17, 24],
                            labels=['nigth', 'morning', 'afternoon', 'evening'],
                            right=False,
                            include_lowest=True)
commits
```

#### ... and sort by 'daytime'

```python
commits_group = commits.groupby('daytime').date.value_counts().unstack(0)
commits_group
```

```python
commits_group.info()
```

### Create a graph (bar chart) with Pandas method .plot()

```python
ax = commits_group.plot(kind='bar',
                   stacked=True,
                   figsize=(15, 8),
                   title='Commits per day',
                   fontsize=8,
                   rot=90,
                   color=['#00008b', '#fffb00', '#ff5500', '#680080'])
ax.set_facecolor('#c2fff4')
ax.grid('on', which='major', axis='y')
```

#### Get sum of the commits for daytime

```python
commits_group.sum()
```

### When do our users usually commit the labs: in the night, morning, afternoon, or evening?
#### The answer is evening and afternoon.

### Which day has the most number of commits and at the same time, the number of commits in the evening is higher than in the afternoon?

```python
commits_group.loc[commits_group['evening'] > commits_group['afternoon']].sum(axis=1).idxmax()
```

#### The answer: 12.05.2020

### Close the connection

```python
connect.close()
print('Connection to SQLite closed')
```
